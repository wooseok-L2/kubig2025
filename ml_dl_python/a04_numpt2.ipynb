```python
a1 = np.fromfunction(lambda x, y, z: x + y + z, (2, 5, 4), dtype=np.int64)
a2 = a1[:, 1::2, :3]  # 얕은 복사

print(a1)

a2[1, ...] = -1

print(a1)

for ar in a2:
    print(ar)
```

```python
a1 = np.fromfunction(lambda x, y, z: x + y + z, (2, 5, 4), dtype=np.int64)
a2 = np.arange(1, a1.size / 2 + 1).reshape(1, 5, 4) * 10
print(a1)
print(a2)
```

```python
a3 = np.vstack((a1[0, ...], a2[0, ...]))
print(a3)
a4 = np.hstack((a1[0, ...], a2[0, ...]))
print(a4)
a5 = np.concatenate((a1, a2), 0)
print(a5)
a6 = np.concatenate((a1[0, ...], a2[0, ...]), 0)  # hstack
a7 = np.concatenate((a1[0, ...], a2[0, ...]), 1)  # vstack
print(a6)
print(a7)
```

```python
# split
a = np.arange(24).reshape(3, 2, 4)
# b1, b2, b3 = np.vsplit(a, 3)  # 얕은 복사
# b4, b5 = np.hsplit(a, 2)  # 얕은 복사
b1 = a[0, ...]
b2 = a[1, ...]
b3 = a[2, ...]
b4 = a[:, 0, :]
b5 = a[:, 1, :]

print(a)
b1[...] = 100
print(b1, b2, b3, b4, b5, sep="\n\n\n")
print(a)
```

```python

a1 = np.arange(1, 5)
a2 = np.arange(1, 7).reshape((2, 3))
a3 = np.linspace(10, 120, 12).reshape((3, 2, 2))
b1 = a1.T
b2 = a2.transpose()
b2[0] = 100
b3 = np.transpose(a3)

print(a1, a2, a3, b1, b2, b3, sep="\n\n\n")
```

```python
a1 = np.array([[2, -1, 5], [-5, 2, 2], [2, 1, 3]])
a2 = np.array([[0, 1, 0], [1, 0, 1], [1, 1, 0]])
a3 = np.array([1, -1, 1])
a4 = np.array([1, 0, 1]).reshape((3, 1))

b1 = np.dot(a1, a2)
b2 = a1.dot(a3)
b3 = a1.dot(a4)

print(a1, a2, a3, a4, b1, b2, b3, sep="\n\n\n")
```
